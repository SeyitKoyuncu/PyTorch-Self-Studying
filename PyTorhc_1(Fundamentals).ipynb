### Random Tensors

Random tensors are important because the way many neural networks learn is that they start with tensors full of random numbers and then adjust those random numbers to better represent the data.

In PyTorch, a random tensor is a multi-dimensional array (tensor) filled with random values. The torch.randn() function is commonly used to create random tensors with elements drawn from a normal distribution with mean 0 and standard deviation 1. The torch.rand() function, on the other hand, creates random tensors with elements drawn from a uniform distribution between 0 and 1.

**A standard deviation (or ):**
* is a measure of how dispersed the data is in relation to the mean. Low standard deviation means data are clustered around the mean, and high standard deviation indicates data are more spread out.

**Random tensors are important in various aspects of machine learning and deep learning. Some of the key reasons why random tensors play a crucial role are as follows:**

  * Initialization of Model Parameters: In many machine learning algorithms, especially in deep learning, model parameters (e.g., weights and biases of neural networks) are initialized with random values. Initializing parameters with random tensors helps break the symmetry and allows the model to start learning from diverse points in the parameter space. Proper initialization can significantly impact the training process and the final performance of the model.

  *  Data Augmentation: Random tensors are used in data augmentation techniques to create variations of training data. For example, in image classification tasks, random rotations, translations, and flips can be applied to the input images using random tensors, leading to a more robust and generalized model.

  *  Dropout Regularization: Dropout is a regularization technique used in deep neural networks to prevent overfitting. During training, random tensors with binary values (0 or 1) are used to mask out a random subset of neurons at each training iteration, effectively introducing noise and promoting robustness.

  * Initialization of Embeddings: In natural language processing tasks, embeddings (dense vector representations of words or tokens) are commonly initialized with random tensors. As the model trains, these embeddings are updated based on the learning task, capturing semantic relationships between words.

  *  Random Sampling: Random tensors are used for random sampling in various contexts, such as generating random noise in generative models like Generative Adversarial Networks (GANs) or sampling from probability distributions in reinforcement learning algorithms.

  *  Exploration in Reinforcement Learning: In reinforcement learning, agents explore the environment by taking random actions with a certain probability. These random actions are often determined using random tensors.

  * Testing and Debugging: Random tensors can also be useful during testing and debugging to simulate various scenarios and assess model performance under different conditions.

### Matrix multipication

Two main ways of performing multipication in neural networks and deep learning:
1. Element-wise multipication
2. Matrix multipication (dot product) (https://www.mathsisfun.com/algebra/matrix-multiplying.html)

There are two main rules that performing matrix multipication need to satisfy:
1. The inner **inner dimensions** must match:
* (3,2) @ (3,2) won't work
* (2,3) @ (3,2) will work
* (3,2) @ (2,3) will work

2. The resulting matrix has the shape of the **outer dimensions**:
* (2, 3) @ (3, 2) -> (2, 2)
* (3, 2) @ (2, 3) -> (3, 3)
* (5, 2) @ (2, 3) -> (5, 3)

http://matrixmultiplication.xyz/

````python
torch.matmul(torch.rand(5,2), torch.rand(2, 3))
````

## Reshaping, stacking, squeezing and unsqueezing tensors

* Reshaping - reshapes an input tensor to a defined shape
* View - Return a view of an input tensor of certain shape but keep the same memory as the original tensor
* Stacking - combine multiple tensors on top of each other (vstack) or side by side (hstack)
* Squeeze - removes all '1' dimensions from a tensor
* Unsqueeze - add a '1' dimension to a target tensor
* Permute - Return a view of the input with dimensions permuted (swapped) in a certain way

-- Reshaping and view is the same things but, their work standarts are different, when used same number of elements you can use view but if you want to manipulate reshaping better, when you use view if you change some element in the view also orginal tensors this element will change

````python
# Lets create a tensor
import torch
x = torch.arange(1., 11.)
print(x)
print(x.shape)
````

````python
# Add an extras dimesion
x_reshaped = x.reshape(5, 2)
print(x_reshaped)
print(x_reshaped.shape)
````

````python
# Change the view
y = torch.arange(1., 10.)
z = y.view(1,9)
print(z)
print(z.shape)
````

````python
# Changing z changes y (because view of a tensor shares the same memory as the original input)
z[:, 0] = 5
print(z)
print(y)
````

````python
# Stack tensors on top of each other
x_stacked = torch.stack([x, x, x, x], dim = 1)
print(x_stacked)
````

````python
# torch.squeeze() - removes all single dimensions from a target tensor
print(f"Previous tensor: {z}")
print(f"Previous shape: {z.shape}")
z_squeeze = z.squeeze()
print(f"New  tensor: {z_squeeze}")
print(f"Previous shape: {z_squeeze.shape}")
````

````python
# torch.unsqueeze() - adds a single dimension to a target tensor at a specific dim
print(f"Previous taget: {z_squeeze}")
print(f"Previous shape: {z_squeeze.shape}")

#add an extra dimension with unsqueeze
z_unsqueezed = z_squeeze.unsqueeze(dim = 0)
print(f"\nNew tensor: {z_unsqueezed}")
print(f"\nNew shape: {z_unsqueezed.shape}")
````

````python
# torch.permute - rearranges the dimensions of a target tensor in a specified order
x_original = torch.rand(size = (224, 223, 3))

# permute the original tensor to rearrange the axis or dim orde
x_permuted = x_original.permute(2, 0, 1) #shifts axis 0 -> 1, 1 -> 2, 2 -> 0

# x_permuted and x_original share the same memory
print(f"Previous shape: {x_original.shape}")
print(f"New shape: {x_permuted.shape}")
````

````python
# evidence of the memory sharing
x_original[0,0,0] = 1234
print(x_permuted[0,0,0])
print(x_original[0,0,0])
````

## PyTorch tensors & numpy

* Data in Numpy, want in PyTorch tensor -> 'torch.from_numpy(ndarray)'
* PyTorch tensor -> NumPy -> 'torch.Tensor.numpy()'

````python
# NumPy array to tensor
import torch
import numpy as np

array = np.arange(1.0, 8.0)
tensor = torch.from_numpy(array) #numpy data type is float64 will you have to convert to the float 32
print(array)
print(tensor)
````

````python
# Change the value of array, what will this do to tensor so dont share memory
array = array + 1
print(array)
print(tensor)
````

````python
# Tensor to NumPy array
tensor = torch.ones(7)
numpy_tensor = tensor.numpy()
print(tensor)
print(numpy_tensor)
````

````python
print(numpy_tensor.dtype)
````

````python
# Change the tensor, what happens to numpy_tensor so dont share memory
tensor = tensor + 1
print(tensor)
print(numpy_tensor)
````

## Reproducibilty (trying to take random out of random)

In short how a neural network learns:
'start with random numbers -> tensor operations -> update random numbers to try and make them better representations of the data -> again -> again -> again...'

To reduce the randomness in neural networks and PyTorch comes the concept of a **random seed**.

Essentially what the random seed does is "flavour" the randomness.

````python
import torch

# Create two random tensors
random_tensor_A = torch.rand(3,4)
random_tensor_B = torch.rand(3,4)

print(random_tensor_A)
print(random_tensor_B)
print(random_tensor_A == random_tensor_B)
````

````python
# Lets's make some random but reproducible tensors
import torch

# Set the random seed
RANDOM_SEED = 42
torch.manual_seed(RANDOM_SEED)
random_tensor_C = torch.rand(3,4)

torch.manual_seed(RANDOM_SEED)
random_tensor_D = torch.rand(3,4)

print(random_tensor_C)
print(random_tensor_D)
print(random_tensor_C == random_tensor_D)
````
